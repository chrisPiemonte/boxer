```python
import sys
from PIL import Image
from pathlib import Path
root_path = Path('../')
images_path = Path(root_path, 'resources/images/siw/')
sys.path.append(str(root_path))

from boxer.nb.widget import SingleImageWidget
```

## SingleImageWidget object and show UI

```python
example_image_path = str(images_path) + '/elon-smoke.jpg'
widget = SingleImageWidget(example_image_path)
widget.show()
```

## Get box coordinates
FORMAT: **<*top*,*left*,*right*,*bottom*,*label*>**

```python
# Get box as BoundingBox object
box = widget.get_box()
print(f'BOX COORDINATES: {box}, TYPE: {type(box)}')
      
# Get box as string
box_string = widget.get_string_box()
print(f'BOX COORDINATES: {box_string}, TYPE: {type(box_string)}')
```
